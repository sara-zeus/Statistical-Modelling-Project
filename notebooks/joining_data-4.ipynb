```python
import sqlite3

connection = sqlite3.connect('POI_and_bikes.db')
cursor = connection.cursor()

# Corrected query with proper syntax
cursor.execute("SELECT categories,SUM([Number of Available Bikes]) FROM final GROUP BY categories ORDER BY SUM([Number of Available Bikes]) DESC LIMIT 10   ")

rows = cursor.fetchall()

for row in rows:
    print(row)

cursor.close()
connection.close()
```

```python
import sqlite3

connection = sqlite3.connect('POI_and_bikes.db')
cursor = connection.cursor()

# Corrected query with proper syntax
cursor.execute("SELECT rating,SUM([Number of Available Bikes]) FROM final GROUP BY categories ORDER BY SUM([Number of Available Bikes]) DESC, rating DESC  LIMIT 10   ")

rows = cursor.fetchall()

for row in rows:
    print(row)

cursor.close()
connection.close()
```
